# Insurance Data Exercise 



**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Read insurance-data.csv as a dataframe called Ins.**

```python
Ins = pd.read_csv("insurance-data.csv")
print(Ins.shape)
```

**Check the head of the DataFrame.**

```python
Ins.head()
```

**Explore The Data Using Pandas Methods**

```python
# Exploring the data using pandas methods :'info', 'columns', 'describe', 'dtype',...
print(f"DataFrame info : {Ins.info()}\n=================================")
print(f"DataFrame columns :\n{Ins.columns}\n=================================")
print(f"The type of each column :\n{Ins.dtypes}\n=================================")
print(f"How much missing value in every column :\n{Ins.isna().sum()}\n=================================")
```

As we can see there is no missing values

*************************************************

**Describing the data**

```python
Ins.describe()
```

*********************************************************

**Smoker Vs Charges**

```python
x=Ins.iloc[:,:5].values
y=Ins.iloc[:,6].values
print(x[:5])
print(y[:5])
```

```python
plt.bar(Ins['smoker'],y,)
plt.xlabel('smoker')
plt.ylabel('Charges')
plt.title('smoker vs charges')
plt.figure()
```

The relation between smoking and having more charges is a Positive relationship That means smokers are more willingly to get more charges

*******************************************************************************************************

**Smokers Vs Non Smokers**

```python
colors = ['black', 'red']
labels = "Smoker", "Non-Smoker"
size = (Ins['smoker'].value_counts())
plt.rcParams['figure.figsize'] = (8, 8)
plt.pie(size, colors = colors, labels = labels, shadow = True)
plt.title('Number of Smoking vs Non-Smoking Members')
plt.show()
```

As we can see the difference between smokers and non smokers is very big

*****************************************************************************

**Age Vs Charges**

```python
plt.figure(figsize = (18, 8))
sns.set(style="whitegrid")
sns.scatterplot(x= Ins['age'] , y= Ins['charges'], color = 'r')
plt.title("Age vs Charges")
sns.despine()
```

As we can see age does affect charges older people have more charges than younger ones.

**********************************************************************

**Males Vs Females**

```python
colors = ['royalblue', 'pink']
labels = "Male", "Female"
size = (Ins['sex'].value_counts())
plt.rcParams['figure.figsize'] = (8, 8)
plt.pie(size, data = Ins, colors = colors, labels = labels, shadow = True)
plt.title('Number of Male vs Female Members')
plt.show()
```

Data shows that the number of males that are interested in insurance are slightly more than women.

*****************************************************************************

**Distribution of Members Living in Different Regions**

```python
plt.figure(figsize = (18, 10))
sns.countplot(Ins['region'], palette='YlOrBr').set_title('Distribution of Members Living in Different Regions')
plt.xlabel('Regions')
plt.ylabel('count')
sns.despine()
plt.show()
```

***********************************************

**Does Where You come From Affect on your Charges?**

```python
plt.figure(figsize = (18, 8))
sns.boxplot(x = 'region', y = 'charges', data = Ins, palette ='YlOrBr')
plt.title('Region vs Charges')
sns.despine()
```
